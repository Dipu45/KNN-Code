```
from google.colab import drive
drive.mount('/content/gdrive')
from numpy import genfromtxt
import numpy as np
from collections import Counter
import random
path='/content/gdrive/MyDrive/AI Lab/dataset/iris.csv'
data=genfromtxt(path,delimiter=',')
data=data.tolist()
#for i in range(5):
 # print(data[i])
Train_set=[] 
Val_set=[]
Test_set=[]
random.shuffle(data)
for s in data:
  R=random.random()
  if R>=0 and R<=0.7:
	  Train_set.append(s) 
  elif R>0.7 and R<=0.85:
  	Val_set.append(s)
  else:
		 Test_set.append(s)
K_accuracy = -1
k_val = -1
for k in range(5,30,5):
    #k=7
    #######for val_set:
    totalNumberOfGuess=0;
    for value in Val_set:
      L=[]
      p1=np.array(value[0:len(value)-1])
      for train in Train_set:
        p2=np.array(train[0:len(train)-1])
        distance= np.linalg.norm(p1-p2)
        temp=[distance,train[len(train)-1]]
        L.append(temp)
        L.sort()

    # print(L)
      newList=[]
      for i in range(k):
        newList.append(L[i][1])
      print(newList)
      major=Counter(newList)
      majority_class=major.most_common()[0][0];
      print(majority_class)
      if majority_class==value[len(value)-1]:
          totalNumberOfGuess+=1
    print('Number of matched data:',totalNumberOfGuess)
    accuracy=(totalNumberOfGuess *100)/len(Val_set)
    print('Validation accuracy:',accuracy,'Number of samples:',k)
    if accuracy>K_accuracy:
      K_accuracy = accuracy
      k_val = k

print("Best K value so far: ",K_accuracy,'Samples required to get best accuracy value',k_val)

###########test_set
totalNumberOfGuess=0;   
for value in Test_set:
  L=[]
  p1=np.array(value[0:len(value)-1])
  for train in Train_set:
   p2=np.array(train[0:len(train)-1])
   distance= np.linalg.norm(p1-p2)
   temp=[distance,train[len(train)-1]]
   L.append(temp)
  L.sort()
 # print(L)
  newList=[]
  for i in range(k_val):
    newList.append(L[i][1])
  print(newList)
  major=Counter(newList)
  majority_class=major.most_common()[0][0];
  print(majority_class)
  if majority_class==value[len(value)-1]:
      totalNumberOfGuess+=1
print(totalNumberOfGuess)
accuracy=(totalNumberOfGuess *100)/len(Test_set)
print('Test Accuracy:',accuracy)
```

```
from google.colab import drive
drive.mount('/content/gdrive')
from numpy import genfromtxt
import numpy as np
from collections import Counter
import random
path='/content/gdrive/MyDrive/AI Lab/dataset/diabetes.csv'
data=genfromtxt(path,delimiter=',')
data=data.tolist()
#for i in range(5):
 # print(data[i])
Train_set=[] 
Val_set=[]
Test_set=[]
random.shuffle(data)
for s in data:
  R=random.random()
  if R>=0 and R<=0.7:
	  Train_set.append(s) 
  elif R>0.7 and R<=0.85:
  	Val_set.append(s)
  else:
		 Test_set.append(s)


K_error = 100000
k_val = 1000000

for k in range(5,30,5):
    #k=7
    #######for val_set:
    error=0;
    for value in Val_set:
      L=[]
      p1=np.array(value[0:len(value)-1])
      for train in Train_set:
        p2=np.array(train[0:len(train)-1])
        distance= np.linalg.norm(p1-p2)
        temp=[distance,train[len(train)-1]]
        L.append(temp)
        L.sort()
    newList=[]
    for i in range(k):
      newList.append(L[i][1])
    print(newList)
    sum=0
    for i in range(k):
     sum=sum+newList[i]
     avg=sum/k
    print('avg',avg)
    error_check=Val_set[i][10]-avg
    error=error+np.square(error_check)
    print('error:',error)
    MSE=(error/len(Val_set))
    print('Mean Squared error:',MSE,'For K=',k)
    if MSE<K_error:
     K_error =  MSE
     k_val = k

print("Best K value so far: ",K_error,'Samples required to get least error value',k_val)

###########test_set  


error=0;
for value in Test_set:
  L=[]
  p1=np.array(value[0:len(value)-1])
  for train in Train_set:
   p2=np.array(train[0:len(train)-1])
   distance= np.linalg.norm(p1-p2)
   temp=[distance,train[len(train)-1]]
   L.append(temp)
  L.sort()
  newList=[]
  for i in range(k_val):
   newList.append(L[i][1])
  print(newList)
  sum=0
  for i in range(k_val):
   sum=sum+newList[i]
   avg=sum/k_val
  print('Avg:',avg)
  error_check=Val_set[i][10]-avg
  error=error+np.square(error_check)
  print('error:',error)
  MSE=(error/len(Test_set))
print('Best Test MSE',MSE,'For K=',k_val)
  

```
